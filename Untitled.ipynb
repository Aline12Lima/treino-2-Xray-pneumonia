```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from collections import Counter
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import random
import os
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
# --- 1. CONFIGURAÇÃO DE DIRETÓRIOS ---
# Estes caminhos devem apontar para onde você moveu as pastas 'train' e 'test'
train_dir = "dataset/chest_xray/train"
test_dir = "dataset/chest_xray/test"

# --- 2. DATA AUGMENTATION E PRÉ-PROCESSAMENTO ---
train_datagen = ImageDataGenerator(
    rescale=1./255,          # Normalização (essencial)
    rotation_range=10,       # Data Augmentation: rotação
    zoom_range=0.1,          # Data Augmentation: zoom
    horizontal_flip=True,    # Data Augmentation: espelhamento
    validation_split=0.2     # Separa 20% para validação
)

# --- 3. CRIAR GERADORES DE DADOS ---
print("--- Preparando Geradores de Dados ---")
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    subset="validation"
)

# Gerador de Teste (apenas normalização, sem Data Augmentation, e shuffle=False)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    shuffle=False
)

print("\nGeradores de dados prontos.")
```

```python
# --- 1. MODELO BASE (Transfer Learning) ---
base_model = DenseNet121(weights="imagenet", include_top=False, input_shape=(224, 224, 3))

# Congelar as camadas base
for layer in base_model.layers:
    layer.trainable = False

# --- 2. CAMADAS DE CLASSIFICAÇÃO (TOPO) ---
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation="sigmoid")(x)
model = Model(inputs=base_model.input, outputs=predictions)

# --- 3. COMPILAÇÃO E BALANCEAMENTO ---
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss="binary_crossentropy",
              metrics=["accuracy", tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()])

# Cálculo de pesos para balanceamento (class_weights)
counter = Counter(train_generator.classes)
max_val = float(max(counter.values()))
class_weights = {class_id: max_val/num_images for class_id, num_images in counter.items()}

print("Pesos das classes para balanceamento:", class_weights)

# --- 4. TREINAMENTO e SALVAMENTO ---
print("\nIniciando Treinamento (15 Épocas)...")
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=15,
    class_weight=class_weights
)

# SALVAR O MODELO (CRUCIAL!)
model.save("modelo_pneumonia_final.keras")
print("\nModelo DenseNet121 treinado e SALVO como 'modelo_pneumonia_final.keras'.")
```

```python
# --- 1. AVALIAÇÃO FINAL NO CONJUNTO DE TESTE ---
test_loss, test_acc, test_prec, test_rec, test_auc = model.evaluate(test_generator)
print(f"Teste - Loss: {test_loss:.4f}, Acc: {test_acc:.4f}, Precision: {test_prec:.4f}, Recall: {test_rec:.4f}, AUC: {test_auc:.4f}")

# --- 2. PLOTAR GRÁFICOS DE ACURÁCIA E PERDA ---
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Acurácia de Treino')
plt.plot(history.history['val_accuracy'], label='Acurácia de Validação')
plt.title('Acurácia do Modelo')
plt.ylabel('Acurácia')
plt.xlabel('Época')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perda de Treino')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda do Modelo')
plt.ylabel('Perda')
plt.xlabel('Época')
plt.legend()
plt.show()
```

```python
# --- CÉLULA 4: COLETA DE FALSOS NEGATIVOS (FN) ---

# 1. FAZER PREVISÕES E OBTER DADOS
# Recarrega o modelo (garante que o modelo salvo seja usado)
try:
    model = load_model("modelo_pneumonia_final.keras")
except:
    pass # Usa o modelo da sessão, se disponível

test_generator.reset()
predictions = model.predict(test_generator, steps=test_generator.n // test_generator.batch_size + 1, verbose=1)

# Conversão das probabilidades para classes binárias (0 ou 1)
y_pred_classes = np.where(predictions.flatten() > 0.5, 1, 0)
y_true = test_generator.classes
file_names = test_generator.filenames

# 2. CRIAR DATAFRAME DE ERROS
df_erros = pd.DataFrame({
    'FileName': file_names,
    'Y_True': y_true, # 0=Normal, 1=Pneumonia
    'Y_Pred': y_pred_classes,
    'Prob_Pneumonia': predictions.flatten()
})

# 3. FILTRAR SOMENTE FALSOS NEGATIVOS (FN)
# FN: Real é PNEUMONIA (1), Modelo previu NORMAL (0)
# Criamos um único DataFrame de erros (df_erros_fn)
df_erros_fn = df_erros[(df_erros['Y_True'] == 1) & (df_erros['Y_Pred'] == 0)].copy()

# --- 4. ORGANIZAR PARA TAXONOMIA MANUAL ---
df_erros_fn['Taxonomia_Manual'] = ''

print(f"Total de Falsos Negativos (FN) para análise: {len(df_erros_fn)}")

print("\n--- INÍCIO DA TAXONOMIA MANUAL: Falsos Negativos (FN) ---")
print("Lista de arquivos para inspeção visual (apenas FNs):")
print(df_erros_fn[['FileName', 'Prob_Pneumonia']].head(50))
```

```python
from sklearn.cluster import KMeans
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Model

# --- CARREGAR O MODELO SALVO ---
try:
    # Garante que o modelo treinado esteja na variável 'model'
    model = load_model("modelo_pneumonia_final.keras")
    print("Modelo 'modelo_pneumonia_final.keras' carregado com sucesso.")
except Exception as e:
    print(f"ERRO CRÍTICO: Não foi possível carregar o modelo salvo: {e}")
    raise NameError("O modelo salvo não foi encontrado ou está corrompido.")

# --- 1. PREPARAR MODELO PARA EXTRAIR EMBEDDINGS (VETORES DE FEATURE) ---
# O Embedding é a saída da camada GlobalAveragePooling2D
embedding_layer = model.layers[-2] 
embedding_model = Model(inputs=model.input, outputs=embedding_layer.output)

print("Modelo de Embedding criado.")

# --- 2. EXTRAIR EMBEDDINGS DE APENAS FALSOS NEGATIVOS (FN) ---
# df_erros_fn deve vir da Célula 4.
df_erros_completos = df_erros_fn.copy() # Usamos APENAS o DataFrame de FNs
error_file_names = df_erros_completos['FileName'].tolist()
print(f"Total de {len(error_file_names)} Falsos Negativos prontos para clusterização.")

# Função para carregar as imagens de erro
def load_error_images(file_names, test_dir):
    images = []
    for file_name in file_names:
        # Nota: 'test_dir' e 'os' devem estar no escopo (importados na Célula 1)
        full_path = os.path.join(test_dir, file_name)
        
        try:
            img = tf.keras.preprocessing.image.load_img(full_path, target_size=(224, 224), color_mode='rgb')
            img = tf.keras.preprocessing.image.img_to_array(img) / 255.0 # Normalização
            images.append(img)
        except FileNotFoundError:
            print(f"Aviso: Arquivo não encontrado em {full_path}. Pulando.")
        
    if not images:
        raise ValueError("Nenhuma imagem de erro foi carregada. Verifique os caminhos.")
        
    return np.array(images)

# Carregar as imagens de erro
error_images = load_error_images(error_file_names, test_dir)

# Gerar os embeddings (vetores de feature)
error_embeddings = embedding_model.predict(error_images, verbose=1)

print(f"Embeddings extraídos: {error_embeddings.shape}")

# --- 3. APLICAR K-MEANS PARA CLUSTERIZAR OS ERROS ---
K = 4 # Número de clusters para análise
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
df_erros_completos['Cluster'] = kmeans.fit_predict(error_embeddings)

# --- 4. RELATÓRIO DE CLUSTERS E LISTAGEM DE ARQUIVOS (PRONTO PARA TAXONOMIA) ---

# 4a. Distribuição de Contagem
cluster_counts = df_erros_completos.groupby(['Cluster']).size().to_frame(name='Total_FNs_no_Cluster')

print("\n--- Distribuição dos Falsos Negativos (FN) por Cluster ---")
print(cluster_counts)

# 4b. Listagem de Arquivos por Cluster (COMPLETA)
print("\n--- LISTAS COMPLETAS DE ARQUIVOS POR CLUSTER (Para Taxonomia) ---")

# Agrupa o DataFrame por Cluster e agrega os FileNames em uma lista para cada grupo
cluster_file_list = df_erros_completos.groupby('Cluster')['FileName'].apply(list)

# Itera sobre os resultados para imprimir a lista completa
for cluster_id, file_list in cluster_file_list.items():
    print(f"\nCluster {cluster_id}: Total de {len(file_list)} Imagens")
    # IMPRIME A LISTA COMPLETA, LINHA POR LINHA
    for file_name in file_list:
        print(f"- {file_name}")

print("\nDataFrame de Falsos Negativos com Clusterização concluído.")
```

```python
# --- CÉLULA 6: ANÁLISE ESTATÍSTICA DA CERTEZA DE ERRO ---

# O df_erros_completos vem da Célula 5

print("\n--- ANÁLISE DA CERTEZA DE ERRO (Probabilidade Média de PNEUMONIA) ---")

# Agrupa por Cluster e calcula a média da probabilidade que o modelo deu
certeza_media = df_erros_completos.groupby('Cluster')['Prob_Pneumonia'].mean().to_frame(name='Prob_Media_PNEUMONIA')

# A probabilidade de erro é a distância de 0.5. Quanto menor o valor, mais certeza o modelo tinha que era NORMAL.
print(certeza_media.sort_values(by='Prob_Media_PNEUMONIA'))

print("\nINTERPRETAÇÃO (Para a Taxonomia):")
print("O Cluster com o MENOR valor de Prob_Media_PNEUMONIA é o que o modelo estava MAIS CERTO de que a imagem era NORMAL.")
```

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

# --- 1. ACESSAR OS CENTROIDES ---
# Os centros dos clusters são vetores de 1024 dimensões (tamanho do Embedding)
# O objeto 'kmeans' deve vir da Célula 5
centroids = kmeans.cluster_centers_

print(f"Centróides extraídos. Formato: {centroids.shape} (4 clusters, 1024 features)")

# --- 2. ANÁLISE QUANTITATIVA DOS CENTRÓIDES ---

# Para análise, vamos focar nos 10 features (índices) que mais se destacam 
# em cada cluster (os valores mais distantes da média de todos os clusters).

# 2.1 Encontrar o Centroide Médio (para comparação)
mean_centroid = np.mean(centroids, axis=0)

print("\n--- ANÁLISE DOS FEATURES MAIS DISTINTIVOS ---")

# 2.2 Iterar sobre cada cluster
for i, centroid in enumerate(centroids):
    # Calcula a diferença absoluta entre o centróide do cluster 'i' e o centróide médio
    diff = np.abs(centroid - mean_centroid)
    
    # Encontra os índices dos 10 valores mais altos nesta diferença
    # Estes são os 10 features (características) que mais tornam este cluster único
    top_10_indices = np.argsort(diff)[-10:][::-1]
    
    print(f"\nCluster {i}: Total de {len(df_erros_completos[df_erros_completos['Cluster'] == i])} FNs")
    print(f"Os 10 features que mais definem o Cluster {i} (maior peso/diferença):")
    
    # Cria um DataFrame para visualizar os índices e seus valores no centróide
    analysis_df = pd.DataFrame({
        'Indice_Feature': top_10_indices,
        'Valor_no_Centroide': centroid[top_10_indices],
        'Diferenca_Absoluta': diff[top_10_indices]
    })
    
    # Adiciona a Probabilidade Média de FN (retirada da Célula 6 anterior, se rodada)
    prob_media_fn = df_erros_completos[df_erros_completos['Cluster'] == i]['Prob_Pneumonia'].mean()
    print(f"Probabilidade Média de FN (Certeza): {prob_media_fn:.4f}")
    
    print(analysis_df)
```
